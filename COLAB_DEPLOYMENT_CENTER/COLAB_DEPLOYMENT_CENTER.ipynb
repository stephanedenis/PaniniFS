```python
# ğŸ¤– LANCEMENT AGENTS AUTONOMES
import subprocess
import threading
import time

def launch_autonomous_agent(script_path):
    """Lance un agent en arriÃ¨re-plan"""
    try:
        result = subprocess.run(['python3', script_path], 
                               capture_output=True, text=True, timeout=300)
        print(f'âœ… Agent {script_path} exÃ©cutÃ©')
        return result.stdout
    except Exception as e:
        print(f'âš ï¸ Erreur agent {script_path}: {e}')
        return None

# Agents disponibles
agents = [
    'autonomous_workflow_doctor.py',
    'nocturnal_autonomous_mission.py',
    'vacation_productive_system.py'
]

print('ğŸš€ Lancement des agents autonomes...')
for agent in agents:
    if os.path.exists(agent):
        print(f'ğŸ¤– Lancement: {agent}')
        output = launch_autonomous_agent(agent)
        if output:
            print(f'   RÃ©sultat: {output[:100]}...')
    else:
        print(f'âš ï¸ Agent non trouvÃ©: {agent}')

print('âœ… SYSTÃˆME AUTONOME OPÃ‰RATIONNEL EN CLOUD')
print('ğŸ¯ EXTERNALISATION RÃ‰USSIE !')
```
